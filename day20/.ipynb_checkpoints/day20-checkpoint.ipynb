```python
# https://adventofcode.com/2021/day/20
import numpy as np
```

```python
def load_input(txt_path):
    with open(txt_path) as f:
        lines = f.read().strip().split('\n')

    # set image data into matrix.
    image = []
    for line in lines[2:]:
        image.append(list(line))

    mapping = np.where(np.array(list(lines[0]))=='#', 1, 0)    
    image = np.where(np.array(image)=='#', 1, 0)
    
    return mapping, image


def bin2image(image_bin):
    x = np.where(image_bin==1, "*", ".")
    for r in x:
        print(''.join(list(r)))
        
        
def enhance(image, mapping, window_size=3, pad_value=0):        
    pad_width = window_size - 1
    image_ = np.pad(image, pad_width, 'constant', constant_values=pad_value)

    r_max, c_max = np.shape(image_)
    image_new = []
    # scan 
    for r in range(1, r_max-pad_width+1):
        for c in range(1, c_max-pad_width+1):
            window = image_[r-1:r+window_size-1, c-1:c+window_size-1]
            window = window.reshape((-1, window_size**2)).flatten()
            bin_str = ''.join(list(window.astype(str)))
            n = int(bin_str, 2)
            #print(n)
            image_new.append(mapping[n])
    image_new = np.array(image_new)

    return image_new.reshape((r_max-pad_width, c_max-pad_width))
```

```python
## part 1&2
txt_path = 'day20.txt'
mapping, image = load_input(txt_path)

n = 50
for i in range(n):
    if mapping[0] == 0:
        pad_value=0
    else:
        pad_value=i%2

    image = enhance(image, mapping, pad_value=pad_value)
    print(f"after {i+1} times enhancing, {np.sum(image)} pixels are lit.")

# display
#bin2image(image_)
```
