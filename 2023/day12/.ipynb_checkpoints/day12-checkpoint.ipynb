```python
def get_data(line):
    springs, numbers = line.split(' ')
    numbers = [int(n) for n in numbers.split(',')]
    return springs, numbers

def minimal_option(numbers):
    springs_ = []
    for i, n in enumerate(numbers):
        if i == len(numbers)-1:
            spring = '#'*n
        else:
            spring = '#'*n + '.'
        springs_.append(spring)
    return ''.join(springs_)
```
