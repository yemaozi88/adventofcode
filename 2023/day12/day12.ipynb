```python
import os
```

```python
is_test = True
if is_test:
    input_txt = 'input1_.txt'
else:
    input_txt = 'input.txt'
with open(input_txt) as f:
    lines = f.read().strip().split('\n')

lines
```

```python
def get_data(line):
    springs, numbers = line.split(' ')
    numbers = [int(n) for n in numbers.split(',')]
    return springs, numbers

def minimal_option(numbers):
    springs_min = []
    for i, n in enumerate(numbers):
        if i == len(numbers)-1:
            spring = '#'*n
        else:
            spring = '#'*n + '.'
        springs_min.append(spring)
    return ''.join(springs_min)
```

```python
line = lines[1]
springs, numbers = get_data(line)
springs_min = minimal_option(numbers)

print(line)
print(f'springs: {springs}')
print(f'numbers: {numbers}')
print(f'minimal_option: {springs_min}')
```
