```python
## load the directry tree
#with open('input_test.txt') as f:
with open('input.txt') as f:
    lines = f.read().strip().split('\n')

ROOT = Node("ROOT" , parent=None)
dir_parent = ROOT
dirlist = []
for file_id, line in enumerate(lines):
    if '$' in line:
        command = line.replace('$ ', '')
        if command == 'ls':
            pass
        elif command == '..':
            dir_parent = dir_parent.parent.name
        elif command.split(' ')[0] == 'cd':
            dir_parent_ = command.split(' ')[1]
            if dir_parent_ == '/':
                dir_parent_ = "ROOT"
            dir_parent = dir_parent_
    else:
        if dir_parent_ =='/':
            dir_parent_ ="ROOT"
        else:
            dir_parent_ = dir_parent
                
        filesize, filename = line.split(' ')
        filename=filename.replace('.', '_')
        if not filesize=='dir':
            command_node = '{} = Node("{}", parent={}, filesize={})'.format(filename, filename, dir_parent_, filesize)
        else:
            dirlist.append(filename)
            command_node = '{} = Node("{}", parent={})'.format(filename, filename, dir_parent_)
        exec(command_node)
        
# check
for pre, fill, node in RenderTree(ROOT):
    print("%s%s" % (pre, node.name))
```
