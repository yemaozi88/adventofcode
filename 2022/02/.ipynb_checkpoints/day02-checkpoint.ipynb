```python
import os

import numpy as np
```

```python
with open('input.txt') as f:
    #lines = f.read().strip().split('\n')
    lines = f.read()

# rock: A/X->1, 
# paper: B/Y->2, 
# scissors: C/Z->3
lines_ = lines.replace('A', '1').replace('B', '2').replace('C', '3')
```

```python
# part 1
lines = lines_.replace('X', '1').replace('Y', '2').replace('Z', '3')
lines = lines.strip().split('\n')
lines
```

```python
def get_outcome(line):
    shape1, shape2 = line.split(' ')
    shape1 = int(shape1) # opponent
    shape2 = int(shape2) # me

    # point_shape
    point_shape = shape2

    # point_win
    # rock1, paper2, scissors3
    result = shape1-shape2
    if result == 0:
        point_win = 3
    elif (result == 1) or (result == -2):
        point_win = 0
    elif (result == -1) or (result == 2):
        point_win = 6
        
    return point_shape, point_win
```

```python
scores = []
for line in lines:
    point_shape, point_win = get_outcome(line)
    scores.append(point_shape + point_win)
np.sum(scores)
```

```python
def get_outcome2(line):
    shape1, shape2_ = line.split(' ')
    shape1 = int(shape1) # opponent
    #shape2 = int(shape2) # me

    # point_shape
    #point_shape = shape2

    if shape1 == 1:
        if shape2_ == 'X':
            shape2 = 3
        elif shape2_ == 'Y':
            shape2 = 1
        elif shape2_ == 'Z':
            shape2 = 2
    elif shape1 == 2:
        if shape2_ == 'X':
            shape2 = 1
        elif shape2_ == 'Y':
            shape2 = 2
        elif shape2_ == 'Z':
            shape2 = 3
    elif shape1 == 3:
        if shape2_ == 'X':
            shape2 = 2
        elif shape2_ == 'Y':
            shape2 = 3
        elif shape2_ == 'Z':
            shape2 = 1

    point_shape = shape2
    if shape2_ == 'X':
        point_win = 0
    elif shape2_ == 'Y':
        point_win = 3
    elif shape2_ == 'Z':
        point_win = 6

    return point_shape, point_win
```

```python
scores = []
for line in lines:
    point_shape, point_win = get_outcome2(line)
    scores.append(point_shape + point_win)
np.sum(scores)
```
