```python
import os
import sys
import importlib

import numpy as np
import pandas as pd

repos_aoc_2025_dir = os.path.dirname(os.path.abspath(os.getcwd()))
```

```python
def load_data(input_txt):
    ranges_list = []
    ingredient_ids = []
    with open(input_txt) as fp:
        line = fp.readline().replace('\n', '')
        start, end = line.split('-')
        range = {'start': int(start), 'end': int(end)}
        ranges_list.append(range)
        while line != '':
            line = line
            line = fp.readline().replace('\n', '')
            try:
                start, end = line.split('-')
                range = {'start': int(start), 'end': int(end)}
                ranges_list.append(range)
            except:
                pass
            
        while True:
            line = fp.readline().replace('\n', '')
            try:
                ingredient_ids.append(int(line))
            except:
                pass
            
            if line == '':
                break
            
    ranges = pd.DataFrame(ranges_list)
    return ranges, ingredient_ids
```

## part 1

```python
input_txt = os.path.join(repos_aoc_2025_dir, "day01", "input.txt")
ranges, ingredient_ids = load_data(input_txt)
#ranges = ranges.sort_values(by="start")
#ranges
```

```python
#for ingredient_id in ingredient_ids:
#    result = is_fresh(ingredient_id, ranges)
#    print(f"ingredient_id:{ingredient_id} --> {result}")
def is_fresh(ingredient_id, ranges):
    df_ = ranges[(ranges['start'] <= ingredient_id) & (ranges['end'] >= ingredient_id)]
    return len(df_)>0

n_fresh = sum([is_fresh(ingredient_id, ranges) for ingredient_id in ingredient_ids])
print(f"Number of the fresh ingredient IDs: {n_fresh}")
```

## part 2

```python
input_txt = os.path.join(repos_aoc_2025_dir, "day05", "input.txt")
ranges, _ = load_data(input_txt)
ranges = ranges.sort_values(by=["start", "end"])
ranges = ranges.drop_duplicates()
# choose only the rows that has correct range.
ranges = ranges[ranges["start"] <= ranges["end"]]

# add n_ingredient between 'start' and 'end.'
ranges["n_ingredients"] = ranges["end"]-ranges["start"]+1
ranges
```

```python
total_ingredients = 0
start_ = 0
end_   = 0
for _, row in ranges.iterrows():
    start = row["start"]
    end   = row["end"]
    n_ingredients = row["n_ingredients"]

    # if row is smaller range than row, skip. 
    if (start >= start_) and (end <= end_):
        pass
    else:
        # if there is no overlap between row and row_...
        if start > end_:
            total_ingredients += n_ingredients
        else:
            total_ingredients += n_ingredients - (end_ - start + 1)

        # update.
        start_ = start
        end_   = end

print(f"according to the fresh ingredient ID ranges, {total_ingredients} to be fresh.")
```
