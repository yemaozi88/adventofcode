```python
import os
import sys
import importlib
import re
import math

import numpy as np
import pandas as pd

repos_aoc_2025_dir = os.path.dirname(os.path.abspath(os.getcwd()))
```

```python
default_operators = ["*", "+"]
```

## part 1

```python
def load_data(input_txt):
    with open(input_txt) as fp:
        lines = fp.readlines()
    # load operators in the last line.
    operators = [op for op in lines[-1].strip() if op in default_operators]
    
    # load matrix.
    n_lines = len(lines)
    X = np.loadtxt(input_txt, max_rows=n_lines-1)
    X = X.astype(int)
    return X, operators
```

```python
input_txt = os.path.join(repos_aoc_2025_dir, "day06", "input.txt")
X, operators = load_data(input_txt)
n_problems = np.shape(X)[1]
assert n_problems==len(operators), 'n_calcuations == len(operators)!'
```

```python
Y = []
for n_problem in range(n_problems):
    x = X[:, n_problem]
    operator = operators[n_problem]
    assert operator in default_operators, f"operator is not in {default_operators}"
    #print(f"[{operator}] {x}")

    if operator == '*':
        y = math.prod(x)
    else:
        y = np.sum(x)
    Y.append(y)

print("the grand total found by adding together all of the answers to the individual problems")
print(np.sum(Y))
```

## part 2

```python
input_txt = os.path.join(repos_aoc_2025_dir, "day06", "input.txt")

with open(input_txt) as fp:
    lines = fp.readlines()
lines = [line.replace('\n', '') for line in lines]
X         = list(map(list, lines[:-1])) # apply list to all element of lines[:,-1] 
X         = np.array(X)
operators = lines[-1]
len_line  = len(operators) 
loc_operators = [i for i in range(len_line) if operators[i] in default_operators]
```

```python
Y = []
for i in range(len(loc_operators)):
    problem_start  = loc_operators[i]
    operator = operators[problem_start]
    
    if i == len(loc_operators)-1:
        problem_length = len_line - loc_operators[i] + 1
    else:
        problem_length = loc_operators[i+1] - loc_operators[i] - 1 # remove ''
    
    x_str = X[:, problem_start:(problem_start+problem_length)]
    x     = [int(''.join(x_str[:, i])) for i in range(x_str.shape[1])]

    if operator == '*':
        y = math.prod(x)
    else:
        y = np.sum(x)
    Y.append(y)
    
    #print(f"{x}({operator})-->{y}")
print(f"sum of the answers to the individual problems")
print(np.sum(Y))
```
