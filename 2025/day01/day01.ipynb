```python
import os
import sys
import importlib

import numpy as np
import pandas as pd

repos_aoc_2025_dir = os.path.dirname(os.path.abspath(os.getcwd()))
#sys.path.append(os.path.join(repos_aoc_2025_dir, 'day01'))
```

```python
def load_data(input_txt):
    ranges_list = []
    ingredient_ids = []
    with open(input_txt) as fp:
        line = fp.readline().replace('\n', '')
        start, end = line.split('-')
        range = {'start': int(start), 'end': int(end)}
        ranges_list.append(range)
        while line != '':
            line = line
            line = fp.readline().replace('\n', '')
            try:
                start, end = line.split('-')
                range = {'start': int(start), 'end': int(end)}
                ranges_list.append(range)
            except:
                pass
            
        while True:
            line = fp.readline().replace('\n', '')
            try:
                ingredient_ids.append(int(line))
            except:
                pass
            
            if line == '':
                break
            
    ranges = pd.DataFrame(ranges_list)
    return ranges, ingredient_ids
```

```python
input_txt = os.path.join(repos_aoc_2025_dir, "day01", "input.txt")
ranges, ingredient_ids = load_data(input_txt)
#ranges = ranges.sort_values(by="start")
#ranges
```
